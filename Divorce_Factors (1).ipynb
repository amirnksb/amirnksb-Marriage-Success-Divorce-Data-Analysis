## Data Dictionary
'age_gap' - The difference in age between the spouses
'education' - The educational level of the spouses
'cultural_similarities' - How similar is the spouses culture with eachother
'common_interests' - common interests between married couple
'religion_compatibility' - how compatible is the religion of the spouses
'relationship_with_the_spouse_family' - relationship with the spouse family
'love' - the feeling of love between marriage spouses
'the_proportion_of_common_genes' - proportion of a common gene taht may be found in married couples
'loyalty' - the spouses quality of being loyal 
'height_ratio' - height ratio between the spouses
'relationships_before_marriage' - relationship engaged before the spouses where married
'socializing_with_opposite_sex' - how social is are the spouses with thier respective opposite sex
'divorce_probability' - the probablity of a divorce happening in a marriage

## Discussion of Findings


From the analysis above we have a weird relationship between age difference and probability of divorce, the graph indicates that having no age difference and having high age difference have a higher probability of divorce meaning the divorce probability is low when there is a little to medium age difference
Next, we have the relationship between education and probability of divorce which has interesting result, the data shows that the more educated the spouses are the higher chance of divorce compared to uneducated ones. 
Next, we have the graph between cultural similarities and probability of divorce, which showed the more culturally similar the spouses are the lower the probability of divorce going on with another similar graph about loyalty and probability of divorce, which showed that the more the loyalty the less the chance of divorce and vice versa
Then, there is a graph on religion compatibility vs chance of divorce and the graph showed somewhat weird pattern. Which led me to dig up more on religion and marriage and I was able to comprehend the graph, it shows that if there is no religion compatibility there is a high chance of divorce but if there is low religion compatibility it is the same as high religion compatibility both resulting in low chance of divorce – this means that if one of the couple are a believer the other one should be to  since it do not matter what the religion is as long as they both believe in something there is a low chance of divorce
Then, we go to another finding that resulted in no pattern which is relationship with the spouse family and chance of divorce showing that there is no correlation on the other hand we found inverse relationship between proportion of common gene with chance of divorce, this is due to a study in biology suggesting, a research team found that when at least one partner had a genetic variation known as the GG genotype within the oxytocin gene receptor, the couple reported significantly greater marital satisfaction and feelings of security within their marriage
Other studies such as socializing with opposite sex and relationship before marriage showed a direct relationship with the chance of divorce and finally , we have the obvious love, the more the love the less the chance of divorce
